```python
#importation de pyspark
from pyspark import SparkContext
from pyspark.sql import SparkSession
```

```python
# If sc is already created, stop it
if 'sc' in locals() and sc:
    sc.stop()

# Create a new SparkContext
sc = SparkContext("local", "MovieRatingsOver4")

# Create a Spark session
spark = SparkSession(sc)
```

```python
# Create a Spark session
spark = SparkSession(sc)
```

```python
# Load the u.data dataset
data = sc.textFile("u.data")
```

```python
# Filter the dataset to include only movie ratings over 4
filtered_data = data.filter(lambda line: int(line.split('\t')[2]) > 4)
```

```python
# Map function to create key-value pairs (movie ID, 1)
mapped_data = filtered_data.map(lambda line: (int(line.split('\t')[1]), 1))
```

```python
# ReduceByKey function to count the number of ratings over 4 for each movie
result = mapped_data.reduceByKey(lambda a, b: a + b)
```

```python
# Display the results
for movie_id, count in result.collect():
    print(f"Movie {movie_id}: {count} ratings over 4")
```

```python
# Stop the Spark context
sc.stop()
```
