```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
# Initialize Spark Session
spark = SparkSession.builder.appName('MovieLens').getOrCreate()
```

```python
# Load the data
data = spark.read.csv('u.data', sep='\t', inferSchema=True, header=False)
```

```python
# Rename the columns
data = data.withColumnRenamed('_c0', 'user_id')\
           .withColumnRenamed('_c1', 'item_id')\
           .withColumnRenamed('_c2', 'rating')\
           .withColumnRenamed('_c3', 'timestamp')
```

```python
# Count the frequency of each value in each column
for column in data.columns:
    data.groupBy(column).count().show()
```

```python
# Calculate the percentage of each value's appearance
total_rows = data.count()
for column in data.columns:
    data.groupBy(column).count().withColumn('percentage', (col('count') / total_rows) * 100).show()
```
